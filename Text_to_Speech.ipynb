```python
import cv2
import matplotlib.pyplot as plt
import pytesseract
from gtts import gTTS
import IPython.display as ipd
```

```python
file_name="Arabic_OCR.png" 
img_cv = cv2.imread(file_name)
plt.figure(figsize=(20, 20))
plt.imshow(img_cv,cmap = 'gray')
plt.title('template')
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
#pytesseract.pytesseract.tesseract_cmd = r'C:/Program Files/Tesseract-OCR/tesseract.exe'
arabic_text = pytesseract.image_to_string(img_cv , lang='ara',config= ".")

print(arabic_text)
```

```python
tts = gTTS(arabic_text, lang='ar')
tts.save('speech.mp3')
```

```python
audio_path = "speech.mp3"

ipd.Audio(audio_path, autoplay=True)
```
